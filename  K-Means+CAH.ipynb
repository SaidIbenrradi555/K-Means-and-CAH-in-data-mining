# 1 Introduction à K-Means clustering


 K-Means clustering est un algorithme de clustering `non supervisé` qui divise un ensemble de données en `k groupes (clusters)` différents. L'objectif est de minimiser la somme des carrés des distances de chaque point de données à son centre de cluster le plus proche.

 L'algorithme K-Means clustering suit une approche itérative pour assigner chaque point de données à l'un des k clusters. La première étape consiste à `initialiser k centres de cluster`. Les centres de cluster initiaux peuvent être choisis de manière aléatoire ou en utilisant une heuristique.

Ensuite, ` pour chaque point de données, l'algorithme trouve le centre de cluster le plus proche et affecte le point de données à ce cluster`. Dans l'étape suivante, `les centres de cluster sont mis à jour en calculant la moyenne des points de données de chaque cluster`. Ce processus d'affectation et de mise à jour des centres de cluster est `répété jusqu'à ce que les centres de cluster convergent.`

# Implémentation en Python


### Étape 1 : Importer les bibliothèques


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

### Étape 2 : Charger les données


### Clustering  selon le : `Longueur des sépales (cm)` et  `Largeur du sépale en centimètres`

```python
iris = load_iris()
data = iris.data[:, :2] # on sélectionne seulement les deux premières colonnes pour une visualisation facile
```

```python
plt.scatter(data[:,0:1],data[:,1:2],c="green")
plt.title('data fleurs Iris')
plt.xlabel('Longueur des sépales (cm)')
plt.ylabel('Largeur des sépales (cm)')
plt.show()
```

### Étape 3 : Initialiser les centres de cluster

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

### Étape 4 : Affecter les points de données aux clusters

```python
clusters = kmeans.fit_predict(data)
```

### Étape 5 : Visualiser les clusters

```python
plt.scatter(data[clusters == 0, 0], data[clusters == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(data[clusters == 1, 0], data[clusters == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolor')
plt.scatter(data[clusters == 2, 0], data[clusters == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 250, c = 'yellow', label = 'Centroids')
plt.title('Clusters de fleurs Iris')
plt.xlabel('Longueur des sépales (cm)')
plt.ylabel('Largeur des sépales (cm)')
plt.legend()
plt.show()
```

# Introduction CAH 
La classification ascendante hiérarchique (CAH) est une méthode de clustering qui consiste à regrouper des données en fonction de leur similarité. La CAH commence par considérer chaque point de données comme un cluster et fusionne ensuite les clusters les plus similaires jusqu'à ce qu'un seul cluster soit obtenu.


Il existe plusieurs bibliothèques Python pour effectuer le clustering CAH, notamment SciPy, scikit-learn .

Dans ce cours, nous utiliserons la bibliothèque SciPy pour effectuer le clustering CAH.

## Préparation des données

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

## Création de la matrice de distances

```python
Z = linkage(X, 'ward')
```

## Visualisation de la dendrogramme

```python
plt.figure(figsize=(12, 6))
plt.title('Dendrogramme CAH')
plt.xlabel('Index de l\'échantillon')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```

## Coupage du dendrogramme

```python
from scipy.cluster.hierarchy import cut_tree

clusters = cut_tree(Z, n_clusters=2)
```

## Visualisation des clusters


```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters[:, 0], cmap='viridis')
plt.title('Clusters Iris')
plt.xlabel('Longueur du sépale')
plt.ylabel('Largeur du sépale')
plt.show()
```
